```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 120  # more than 1000 rows

data = pd.DataFrame({
    "Order_ID": np.arange(1, n+1),
    "Category": np.random.choice(["Electronics", "Clothing", "Furniture"], n),
    "Region": np.random.choice(["North", "South", "East", "West"], n),
    "Sales": np.random.randint(100, 10000, n),
    "Profit": np.random.randint(-500, 3000, n),
    "Discount": np.round(np.random.uniform(0, 0.5, n), 2),
    "Quantity": np.random.randint(1, 10, n)
})

print(data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.setstyle="whitegrid"
```

```python
sns.lineplot(x="Region", y="Sales", data=data)
plt.title("Sales Trend")
plt.show()
```

```python
sns.barplot(x="Category", y="Sales", data=data)
plt.title("Average Sales by Category")
plt.show()
```

```python
sns.countplot(x="Region", data=data)
plt.title("Number of Orders per Region")
plt.show()
```

```python
sns.histplot(data["Sales"], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
sns.boxplot(x="Category", y="Profit", data=data)
plt.title("Profit Distribution by Category")
plt.show()
```

```python
sns.violinplot(x="Category", y="Sales", data=data)
plt.title("Sales Distribution by Category")
plt.show()
```

```python
sns.scatterplot(x="Sales", y="Profit", data=data, hue="Category")
plt.title("Sales vs Profit")
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
corr = data.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.stripplot(x="Category", y="Sales", data=data)
plt.title("Sales Distribution (Strip)")
plt.show()
```

```python
sns.swarmplot(x="Category", y="Sales", data=data)
plt.title("Sales Distribution (Swarm)")
plt.show()
```

```python
sns.jointplot(x="Sales", y="Profit", data=data, kind="scatter")
plt.show()
```

```python
g = sns.FacetGrid(data, col="Category")
g.map(sns.histplot, "Sales")
plt.show()
```
