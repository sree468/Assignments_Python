```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("tips")
df.head()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time')
plt.subplot(1,2,2)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex')
```

```python
sns.lineplot(data=df, x='total_bill', y='tip')
plt.show()
```

```python
sns.barplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
df.shape
```

```python
sns.countplot(data=df, x='day')
plt.show()
```

```python
sns.boxplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
sns.boxplot(data=df, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
sns.violinplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
sns.violinplot(data=df, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
sns.stripplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
df=df.sample(10000,replace=True)
df.shape
```

```python
sns.regplot(data=df, x='total_bill', y='tip')
plt.show()
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='scatter')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```
