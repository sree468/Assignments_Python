```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Employee_ID": range(1, 101),
    "Age": np.random.randint(21, 60, 100),
    "Salary": np.random.randint(20000, 120000, 100),
    "Experience": np.random.randint(1, 20, 100),
    "Department": np.random.choice(["IT", "HR", "Finance", "Sales"], 100),
    "Performance_Score": np.random.randint(1, 5, 100)
}

df = pd.DataFrame(data)
print(df.head())
```

```python
df
```

```python

# 1. Line Plot
plt.figure(figsize=(20,5))
plt.plot(df["Employee_ID"], df["Salary"],color='#006000')
plt.xlabel("Emplpoyee id")
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```

```python

# 1. Line Plot
plt.figure(figsize=(20,5))
plt.plot(df["Employee_ID"], df["Salary"],linewidth=10,alpha=0.1,markersize=25,color='Blue',marker='o',markerfacecolor='w',
         markeredgecolor='b')
plt.xlabel("Emplpoyee id")
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(df["Employee_ID"], df["Salary"])
plt.subplot(2,2,2)
plt.plot(df["Employee_ID"], 1/df["Salary"],color='r')
plt.subplot(2,2,3)
plt.plot(df["Employee_ID"], 1/df["Salary"],color='b')
plt.subplot(2,2,4)
plt.plot(df["Employee_ID"], 1/df["Salary"],color='g')
```

```python
# 2. Scatter Plot
plt.figure()
plt.scatter(df["Age"], df["Salary"],color='r')
plt.title("Age vs Salary")
plt.show()
```

```python
# 3. Histogram
plt.figure()
plt.hist(df["Salary"])
plt.title("Salary Distribution")
plt.show()
```
