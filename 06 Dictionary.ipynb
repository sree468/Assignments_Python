```python
#Create a dictionary with class values: {'list': list, 'dict': dict, 'set': set}
dict27={'list': list, 'dict': dict, 'set': set}
print(dict27)
```

```python
# Create instances using class dictionary
class_dict = {
    "list": list,
    "dict": dict,
    "set": set,
    "tuple": tuple
}
instances = {}

for name, cls in class_dict.items():
    instances[name] = cls()   # create instance of each class
print(instances)
```

```python
# Create a dictionary with nested sets: {'groups': {{1, 2, 3}, {4, 5, 6}}, 'categories': {{'a', 'b'}, {'c', 'd'}}}
dict49 = {
    'groups': {frozenset({1, 2, 3}), frozenset({4, 5, 6})},
    'categories': {frozenset({'a', 'b'}), frozenset({'c', 'd'})}
}
print(dict49)
```

```python
#Find union of all nested sets
dict50= {
    'groups': {frozenset({1, 2, 3}), frozenset({4, 5, 6})},
    'categories': {frozenset({'a', 'b'}), frozenset({'c', 'd'})}
}
def union_of_nested_sets(container):
    result = set()
    for s in container:
        result |= s
    return result
print(union_of_nested_sets(dict50['groups']))
print(union_of_nested_sets(dict50['categories']))
```

```python
#Create a dictionary with nested lists: {'matrix': [[1, 2], [3, 4]], 'vector': [5, 6, 7]}
dict43={'matrix': [[1, 2], [3, 4]], 'vector': [5, 6, 7]}
print(dict43)
```

```python
#Find sum of each nested list
dict44={'matrix': [[1, 2], [3, 4]], 'vector': [5, 6, 7]}
total = 0
# matrix
for row in dict44['matrix']:
    total += sum(row)
# vector
total += sum(dict44['vector'])
print("Total sum of all numbers:", total)
```
