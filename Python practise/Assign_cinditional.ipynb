```python
'''Leap Year Checker Write a program to check if a year is a leap year:
Divisible by 4 AND (not divisible by 100 OR divisible by 400)
'''
year=2026
if year%4==0 and (year%100!=0 or year%400==0 ):
    print("leap year")
else:
    print("not a leap year")
```

```python
'''Login System with Attempts Create a login system that:
Checks username and password'''
username="sree"
password="sree@1"
user_login=input("enter username")
passw=input("enter the password")
attempts=0
if username==user_login:
    print("correct username")
    if password!=passw:
        attempts+=1
        print(f"attempts:{attempts}")
    else:
        password==passw
        print("correct password")
else:
    attempts+=1
    print(f"attempts:{attempts}")
    
    
    
```

```python
username = "sree"
password = "sree@1"

attempts = 0
max_attempts = 3

while attempts < max_attempts:
    user_login = input("Enter username: ")
    passw = input("Enter password: ")

    if user_login == username and passw == password:
        print("Login successful")
        break
    else:
        attempts += 1
        print(f"atempt count: {max_attempts - attempts}")

if attempts == max_attempts:
    print("Account locked. Too many failed attempts")
```
