```python
#1. Write a function that takes a number as input and returns whether it is even or odd.
def check(n):
    if n%2==0:
        return "Even"
    else:
        return "odd"
a=check(13)
a
```

```python
# Write a function that takes a number and returns its factorial.
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
b=factorial(5)
b
```

```python
#Write a function that takes a string and returns the reversed string.
def rev(str):
    return str[::-1]
c=rev("Hi")
c
```

```python
#4. Write a function that checks whether a number is prime.
def primecheck(n):
    if n<1:
        return "not prime"
    for i in range(2,int((n)**0.5)+1):
        if n%2==0:
            return "not prime"
        else:
            return "prime"
d=primecheck(12)
d
```

```python
#Write a function that takes a list of numbers and returns the largest number.
def large(lst):
    largest=lst[0]
    for i in lst:
        if i>largest:
            largest=i
    return largest
e=large([1,23,3])
e


```

```python
# Write a function that takes a list and returns the sum of all elements without using the built-in `sum()` function.
def add(lst1):
    sum=0
    for i in lst1:
        sum+=i
    return sum
f=add([1,2,3])
f
```

```python
#Write a function that counts the number of vowels in a string.
def vowels(str):
    vo='aeiouAEIOU'
    count=0
    for i in str:
        if i in vo:
            count+=1
    return count
g=vowels('aest')
g
```

```python
#Write a function that checks whether a string is a palindrome.
def pali(str):
    if str==str[::-1]:
        return "palindrome"
    else:
        return "Not palindrome"
h=pali("madam")
h
```

```python
#Write a function that takes two numbers and returns their greatest common divisor (GCD).
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
i = gcd(2, 3)
print(i)
```

```python
#Write a function that removes duplicates from a list.
def dupli(lst2):
    ans=[]
    for i in lst2:
        if i not in ans:
            ans.append(i)
    return ans
j=dupli([1,2,1,3,4])
j
```

```python
#Write a function that takes a list of numbers and returns a new list containing only even numbers.
def even(lst2):
    onlyeven=[]
    for i in lst2:
        if i%2==0:
            onlyeven.append(i)
    return onlyeven
j=even([1,2,1,3,4])
j
```

```python
#13. Write a function that finds the frequency of each character in a string.
def fre(str):
    freq={}
    for i in str:
        freq[i]=freq.get(i,0)+1
    return freq
k=fre("sree")
k

```

```python
#14. Write a function that swaps two numbers without using a third variable.
def swap(a,b):
    a=a+b
    b=a-b
    a=a-b
    return a,b
l=swap(3,4)
l
```

```python
# Write a function that converts a temperature from Celsius to Fahrenheit.
def fah(c):
    fahr=(c*(9/5))+32
    return fahr
m=fah(34)
m

```

```python
#Write a function that returns the Fibonacci sequence up to `n` terms.
def fibonacci(n):
    fib = []
    a, b = 0, 1
    for i in range(n):
        fib.append(a)
        a, b = b, a + b
    return fib
n=fibonacci(5)
n
```
