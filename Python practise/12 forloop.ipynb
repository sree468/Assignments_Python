```python
#Reverse a number using only a for loop (no slicing).
number='12345'
rev=''
for i in number:
    rev=i+rev
print(f"the reverse of {number} is {rev}")
```

```python
#Check whether a number is prime using a for loop.
n = 11
if n <= 1:
    print(False)
else:
    is_prime = True  # Flag variable
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break
    print(is_prime)
        
```

```python
"""
11. From 1 to 50, find:
a) sum of even numbers
b) sum of odd numbers"""
even=0
odd=0
for i in range(1,51):
    if i%2==0:
        even+=i
    else:
        odd+=i
print(f"sum of odd numbers is {odd}")
print(f"sum of even numbers is {even}")
```

```python
#14. Find duplicate values in a list.
dupli= [1,2,3,2,4,5,1]
duplicates=[]
for i in range(len(dupli)):
    for j in range(i+1,len(dupli)):
        if dupli[i]==dupli[j] and dupli[i] not in duplicates:
            duplicates.append(dupli[i])
print(f"duplicates are {duplicates}")
```

```python
#Print the frequency of each character in a string.
text = 'banana'
freq = {}
for ch in text:
    if ch in freq:
        freq[ch]+=1
    else:
        freq[ch]=1
for key,value in freq.items():
    print(f"{key}:{value}")
```
