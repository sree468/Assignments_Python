```python
import re

text = "car"
pattern = r"[arn]"

matches = re.findall(pattern, text)
print("Matches:", matches)
```

```python
text = "hello"
pattern = r"[a-n]"

matches = re.findall(pattern, text)
print("Matches:", matches)
```

```python
text = "2019"
pattern = r"[0123]"

matches = re.findall(pattern, text)
print("Matches:", matches)
```

```python
text = "Room 42"
pattern = r"[0-9]"

matches = re.findall(pattern, text)
print("Matches:", matches)
```

```python
text = "12, 67, 45"
pattern = r"[0-5][0-9]"

matches = re.findall(pattern, text)
print("Matches:", matches)
```

```python
text = "Hello123!"
pattern = r"[a-zA-Z]"

matches = re.findall(pattern, text)
print("Matches:", matches)
```

```python
text = "5+3=8"
pattern = r"[+]"

matches = re.findall(pattern, text)
print("Matches:", matches)
```

```python
import re

text = "café"
print(re.findall(r"\w+", text, re.ASCII))   # only ASCII chars
```

```python
import re

text = "hello HELLO HeLLo"
print(re.findall(r"hello", text))
```

```python
import re

text = "hello HELLO HeLLo"
print(re.findall(r"hello", text, re.I))
```

```python
import re

text = "Hello\nWorld"
print(re.findall(r"^World", text))
```

```python
import re

text = "Hello\nWorld"
print(re.findall(r"^World", text, re.M))
```
