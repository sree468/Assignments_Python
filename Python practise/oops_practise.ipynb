```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.transactions = []   # To store transaction history

    def deposit(self, amount):
        if amount <= 0:
            print("Amount must be positive")
            return
        self.balance += amount
        self.transactions.append(f"Deposited: {amount}")
        print(f"{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Amount must be positive")
            return
        if amount <= self.balance:
            self.balance -= amount
            self.transactions.append(f"Withdrawn: {amount}")
            print(f"{amount} withdrawn successfully.")
        else:
            print("Insufficient balance")

    def display_balance(self):
        print(f"\nAccount Holder: {self.name}")
        print(f"Current Balance: {self.balance}")

    def show_transactions(self):
        print("\nTransaction History:")
        if not self.transactions:
            print("No transactions yet.")
        else:
            for t in self.transactions:
                print(t)


# ---- Main Program ----
name = input("Enter account holder name: ")
balance = float(input("Enter initial balance: "))

account = BankAccount(name, balance)

while True:
    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Display Balance")
    print("4. Show Transactions")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        amount = float(input("Enter amount to deposit: "))
        account.deposit(amount)

    elif choice == '2':
        amount = float(input("Enter amount to withdraw: "))
        account.withdraw(amount)

    elif choice == '3':
        account.display_balance()

    elif choice == '4':
        account.show_transactions()

    elif choice == '5':
        print("Thank you for banking!")
        break

    else:
        print("Invalid choice. Try again.")
```

```python
#Student Management System
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks

    def calculate_total(self):
        return sum(self.marks.values())

    def calculate_average(self):
        if len(self.marks) == 0:
            return 0
        return self.calculate_total() / len(self.marks)

    def get_grade(self):
        avg = self.calculate_average()
        if avg >= 90:
            return "Grade A"
        elif avg >= 75:
            return "Grade B"
        elif avg >= 50:
            return "Grade C"
        else:
            return "Fail"


# Example usage
marks_dict = {
    "Math": 95,
    "Science": 85,
    "English": 78
}

student1 = Student("Sree", marks_dict)

print("Total Marks:", student1.calculate_total())
print("Average Marks:", student1.calculate_average())
print("Grade:", student1.get_grade())
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks   # dictionary {subject: marks}

    def calculate_total(self):
        return sum(self.marks.values())

    def calculate_average(self):
        if len(self.marks) == 0:
            return 0
        return self.calculate_total() / len(self.marks)

    def get_grade(self):
        if len(self.marks) == 0:
            return "No Marks"

        avg = self.calculate_average()
        if avg >= 90:
            return "Grade A"
        elif avg >= 75:
            return "Grade B"
        elif avg >= 50:
            return "Grade C"
        else:
            return "Fail"


# ---- Multiple Students ----
students = []

# Adding student objects
students.append(Student("Sree", {"Math": 95, "Science": 85, "English": 78}))
students.append(Student("Ravi", {"Math": 45, "Science": 50, "English": 55}))
students.append(Student("Anu", {}))  # Empty marks case

# ---- Display Details ----
for student in students:
    print("\nStudent Name:", student.name)
    
    if len(student.marks) == 0:
        print("No marks available")
    else:
        print("Total Marks:", student.calculate_total())
        print("Average Marks:", student.calculate_average())
    
    print("Grade:", student.get_grade())
```

```python
# Library Management system
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True   # default value


class Library:
    def __init__(self):
        self.books = []   # list to store books

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added to library.")

    def borrow_book(self, title):
        for book in self.books:   # loop to search
            if book.title.lower() == title.lower():
                if book.is_available:   # check availability
                    book.is_available = False
                    print(f"You borrowed '{book.title}'.")
                else:
                    print(f"'{book.title}' is already borrowed.")
                return
        print("Book not found")   # if no match

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if not book.is_available:
                    book.is_available = True
                    print(f"You returned '{book.title}'.")
                else:
                    print(f"'{book.title}' was not borrowed.")
                return
        print("Book not found")

    def display_books(self):
        if not self.books:
            print("Library is empty")
            return

        print("\nLibrary Books:")
        for book in self.books:
            status = "Available" if book.is_available else "Not Available"
            print(f"Title: {book.title}, Author: {book.author}, Status: {status}")


# ---- Example Usage ----
lib = Library()

# Adding books
lib.add_book(Book("Python Basics", "John"))
lib.add_book(Book("Data Science", "Alice"))
lib.add_book(Book("Machine Learning", "Bob"))

# Display books
lib.display_books()

# Borrow book
lib.borrow_book("Python Basics")
lib.borrow_book("Python Basics")   # already borrowed
lib.borrow_book("Java")            # not found

# Return book
lib.return_book("Python Basics")
lib.return_book("Python Basics")   # already returned

# Final display
lib.display_books()
```
