```python
def calculate_compound_interest(principal, annual_rate, years, compounds_per_year):
    rate_decimal = annual_rate / 100.0
    amount = principal * ((1 + rate_decimal / compounds_per_year) ** (compounds_per_year * years))
    interest = amount - principal
    return amount, interest

P = 10000
R = 5
T = 2
N = 4

final_amount, compound_interest = calculate_compound_interest(P, R, T, N)

print(f"Principal Amount: ${P:,.2f}")
print(f"Annual Rate: {R}%")
print(f"Time in Years: {T}")
print(f"Compounding Frequency (per year): {N}")
print("-" * 30)
print(f"Total Amount after {T} years: ${final_amount:,.2f}")
print(f"Compound Interest Earned: ${compound_interest:,.2f}")
```

```python
# rot13 encoding
str="hello world"
result = ""
for ch in str:
    if 'a' <= ch <= 'z':
        result += chr((ord(ch) - ord('a') + 13) % 26 + ord('a'))
    elif 'A' <= ch <= 'Z':
        result += chr((ord(ch) - ord('A') + 13) % 26 + ord('A'))
    else:
        result += ch
print(result)
```

```python
def pig_latin(word):
    vowels = "aeiouAEIOU"
    if word[0] in vowels:
        return word + "way"
    else:
        for i, ch in enumerate(word):
            if ch in vowels:
                return word[i:] + word[:i] + "ay"
        return word + "ay"  # fallback if no vowel

sentence = "Pig Latin is fun"
result = ' '.join(pig_latin(w) for w in sentence.split())
print(result)

```
