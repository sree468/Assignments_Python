```python
#Question 1: Calculate the product of first 10 natural numbers
# Your code here
#n=int(input('enter'))
n=10
ans=1
for i in range(1,n+1):
    ans=ans*i
print(f"product of first {n} natural numbers is {ans}".format(n,ans))
```

```python

# Question 2: Find the remainder when 156 is divided by 7
# Your code here
div=156
divisor=7
ans=div/divisor
print(f"{div} is divided by {divisor} is {ans}".format(div,divisor,ans))
```

```python
# Question 3: Calculate the square of 25
# Your code here
import math
n=25
ans=math.sqrt(n)
print(f"sqrt of {n} is {ans}".format(n,ans))
```

```python
# Question 4: Find the cube root of 125
# Your code here
#n=int(input("enter"))
n=125
ans=n**(1/3)
print(f"The cube root of {n} is {ans}".format(n,ans))
```

```python
#Question 5: Calculate the sum of digits in number 12345
# Your code here
num = 12345
total = 0
while num > 0:
    digit = num % 10
    total += digit
    num //= 10
print(f" sum of digits of {num} is {total}".format(num,total))
```

```python
#Question 7: Find the factorial of 8
# Your code here
import math
n=8
ans=math.factorial(n)
print(f"The factorial of {n} is {ans}".format(n,ans))
```

```python
# Question 8: Calculate the average of numbers: 15, 23, 31, 42, 56
# Your code here
#nums = list(map(int, input("Enter numbers: ").split(',')))
nums= 15, 23, 31, 42, 56
length=len(nums)
ans=sum(nums)/length
print(f"The avg of {nums} is {ans}".format(nums,ans))
```
