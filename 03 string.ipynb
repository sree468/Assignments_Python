```python
s = "mississippi"
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
most_char = max(freq, key=freq.get)

print("Most frequent character:", most_char)
print("Count:", freq[most_char])
```

```python
# rot13 encoding
str="hello world"
result = ""
for ch in str:
    if 'a' <= ch <= 'z':
        result += chr((ord(ch) - ord('a') + 13) % 26 + ord('a'))
    elif 'A' <= ch <= 'Z':
        result += chr((ord(ch) - ord('A') + 13) % 26 + ord('A'))
    else:
        result += ch
print(result)
```

```python
def pig_latin(word):
    vowels = "aeiouAEIOU"
    if word[0] in vowels:
        return word + "way"
    else:
        for i, ch in enumerate(word):
            if ch in vowels:
                return word[i:] + word[:i] + "ay"
        return word + "ay"  # fallback if no vowel

sentence = "Pig Latin is fun"
result = ' '.join(pig_latin(w) for w in sentence.split())
print(result)
```
